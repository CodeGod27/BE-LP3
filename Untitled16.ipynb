```python
def fibonacci_recursive_with_step(n, step_count=0):
    if n == 0:
        return [0], step_count
    elif n == 1:
        return [0, 1], step_count
    else:
        fib_n_1, step_count = fibonacci_recursive_with_step(n - 1, step_count)
        fib_n_2, step_count = fibonacci_recursive_with_step(n - 2, step_count)
        step_count += 1

        fib_n = fib_n_1 + [fib_n_1[-1] + fib_n_2[-1]]
        return fib_n, step_count

n = int(input("Enter Number of Terms in the Fibonacci series: "))

fibonacci_series, step_count = fibonacci_recursive_with_step(n - 1)

print("Recursive Fibonacci:", *fibonacci_series)
print(f"Step count: {step_count}")
```

```python
import heapq

class node:
  def __init__(self, freq, symbol, left=None, right=None):
    self.freq = freq
    self.symbol = symbol
    self.left = left
    self.right = right
    self.huff = ''

  def __lt__(self, nxt):
    return self.freq < nxt.freq

# utility function to print huffman codes for all symbols in the newly created Huffman tree
def printNodes(node, val=''):
  newVal = val + str(node.huff)
  if node.left:
    printNodes(node.left, newVal)
  if node.right:
    printNodes(node.right, newVal)
  if not node.left and not node.right:
    print(f"{node.symbol} -> {newVal}")

# calculate fibonacci series
def fibonacci(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n - 1) + fibonacci(n - 2)

# generate fibonacci series up to n terms
n = 10
fibonacci_series = [fibonacci(i) for i in range(n)]

# create frequency table from fibonacci series
freq_table = {}
for num in fibonacci_series:
  if num in freq_table:
    freq_table[num] += 1
  else:
    freq_table[num] = 1

# convert frequency table to list of nodes
nodes = []
for num, freq in freq_table.items():
  nodes.append(node(freq, str(num)))

# build Huffman tree
while len(nodes) > 1:
  left = heapq.heappop(nodes)
  right = heapq.heappop(nodes)
  left.huff = 0
  right.huff = 1
  newNode = node(left.freq + right.freq, left.symbol + right.symbol, left, right)
  heapq.heappush(nodes, newNode)

# print Huffman codes
printNodes(nodes[0])
```

```python
def knapsack_branch_and_bound(items, capacity):
    # Sort items in decreasing order of value per weight ratio
    items.sort(key=lambda item: item['value'] / item['weight'], reverse=True)

    # Create an empty queue
    queue = []

    # Create a dummy node of the decision tree and enqueue it to the queue
    dummy_node = {'value': 0, 'weight': 0, 'level': 0}
    queue.append(dummy_node)

    # Initialize the maximum value found so far
    max_value = 0

    # Explore the decision tree using breadth-first search
    while queue:
        # Get the current node from the queue
        current_node = queue.pop(0)

        # Check if the current node is a leaf node (no more items to include)
        if current_node['level'] == len(items):
            # Update the maximum value if necessary
            max_value = max(max_value, current_node['value'])
            continue

        # Create a child node for including the current item
        include_node = {
            'value': current_node['value'] + items[current_node['level']]['value'],
            'weight': current_node['weight'] + items[current_node['level']]['weight'],
            'level': current_node['level'] + 1
        }

        # Compute the upper bound for the include node
        remaining_capacity = capacity - include_node['weight']
        remaining_items = [item for item in items[include_node['level']:]]
        remaining_items.sort(key=lambda item: item['value'] / item['weight'], reverse=True)
        upper_bound_include = include_node['value'] + sum([item['value'] for item in remaining_items[:min(len(remaining_items), remaining_capacity)]])

        # Add the include node to the queue if it's promising
        if upper_bound_include > max_value:
            queue.append(include_node)

        # Create a child node for excluding the current item
        exclude_node = {
            'value': current_node['value'],
            'weight': current_node['weight'],
            'level': current_node['level'] + 1
        }

        # Add the exclude node to the queue
        queue.append(exclude_node)

    return max_value

# Example usage
if __name__ == "__main__":
    items = [
        {'name': 'item1', 'value': 60, 'weight': 10},
        {'name': 'item2', 'value': 100, 'weight': 20},
        {'name': 'item3', 'value': 120, 'weight': 30}
    ]
    max_capacity = 50

    max_value = knapsack_branch_and_bound(items, max_capacity)
    print("Maximum value in the knapsack:", max_value)
```

```python
def is_safe(board, row, col):
    for i in range(col):
        if board[row][i] == 1:
            return False
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    for i, j in zip(range(row, len(board), 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    return True

def solve_n_queens(board, col):
    if col == len(board):
        return True
    for i in range(len(board)):
        if is_safe(board, i, col):
            board[i][col] = 1
            if solve_n_queens(board, col + 1):
                return True
            board[i][col] = 0  # Backtrack if placement leads to no solution
    return False

def print_board(board):
    for row in board:
        print(" ".join(["Q" if cell == 1 else "-" for cell in row]))

# Create an 8x8 chessboard with all cells initialized to 0
n = 9
chessboard = [[0 for _ in range(n)] for _ in range(n)]
# Place the first queen in the first row, first column
chessboard[0][0] = 1
# Try to solve the remaining queens
if solve_n_queens(chessboard, 1):
    print("Solution found:")
    print_board(chessboard)
else:
    print("No solution exists.")
```

```python
import random
import time


def quick_sort(array, low, high):
    if low < high:
        pivot = partition(array, low, high)
        quick_sort(array, low, pivot - 1)
        quick_sort(array, pivot + 1, high)


def randomized_quick_sort(array, low, high):
    if low < high:
        pivot = random_partition(array, low, high)
        randomized_quick_sort(array, low, pivot - 1)
        randomized_quick_sort(array, pivot + 1, high)


def partition(array, low, high):
    pivot = array[high]
    i = low - 1

    for j in range(low, high):
        if array[j] <= pivot:
            i += 1
            array[i], array[j] = array[j], array[i]

    array[i + 1], array[high] = array[high], array[i + 1]
    return i + 1


def random_partition(array, low, high):
    random_pivot = random.randint(low, high)
    array[random_pivot], array[high] = array[high], array[random_pivot]
    return partition(array, low, high)


def analyze_quicksort(array, sort_type):
    start_time = time.time()
    if sort_type == "deterministic":
        quick_sort(array, 0, len(array) - 1)
    elif sort_type == "randomized":
        randomized_quick_sort(array, 0, len(array) - 1)
    end_time = time.time()

    execution_time = end_time - start_time
    print(f"Execution time for {sort_type} quicksort: {execution_time} seconds")


if __name__ == "__main__":
    array = [5, 2, 4, 1, 3]

    print("Original array:", array)

    analyze_quicksort(array.copy(), "deterministic")
    analyze_quicksort(array.copy(), "randomized")
```

```python
def merge(left, right):
    # Create an empty merged list
    merged = []

    # Initialize two pointers to the starting elements of the left and right lists
    i, j = 0, 0

    # Compare elements from both lists and add the smaller element to the merged list
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    # Add any remaining elements from the left or right list to the merged list
    merged.extend(left[i:])
    merged.extend(right[j:])

    # Return the merged list
    return merged


def merge_sort(array):
    if len(array) <= 1:
        return array

    # Divide the array into two halves
    mid = len(array) // 2
    left = array[:mid]
    right = array[mid:]

    # Recursively sort the left and right halves
    left = merge_sort(left)
    right = merge_sort(right)

    # Merge the sorted left and right halves
    merged = merge(left, right)

    # Return the merged and sorted array
    return merged
array = [5, 2, 4, 1, 3]
sorted_array = merge_sort(array)
print(sorted_array)

```

```python
def selection_sort(array):
    for i in range(len(array)):
        min_index = i
        for j in range(i + 1, len(array)):
            if array[j] < array[min_index]:
                min_index = j

        array[i], array[min_index] = array[min_index], array[i]

    return array
array = [5, 2, 4, 1, 3]
sorted_array = selection_sort(array)
print(sorted_array)
```

```python
def insertion_sort(array):
    for i in range(1, len(array)):
        key = array[i]
        j = i - 1

        while j >= 0 and array[j] > key:
            array[j + 1] = array[j]
            j -= 1

        array[j + 1] = key

    return array
array = [5, 2, 4, 1, 3]
sorted_array = insertion_sort(array)
print(sorted_array)
```

```python

def heapify(array, n, i):
    parent = (i - 1) // 2
    largest = i

    if parent >= 0 and array[parent] > array[largest]:
        largest = parent

    if parent != largest:
        array[parent], array[largest] = array[largest], array[parent]
        heapify(array, n, parent)


def build_heap(array, n):
    for i in range(n // 2, -1, -1):
        heapify(array, n, i)


def heap_sort(array):
    n = len(array)
    build_heap(array, n)

    for i in range(n - 1, 0, -1):
        array[i], array[0] = array[0], array[i]
        heapify(array, i, 0)

    return array
array = [5, 2, 4, 1, 3]
sorted_array = heap_sort(array)
print(sorted_array)
```
